#### Resizing the image

```python
plt.imshow(fix_img)
```

```python
fix_img.shape
```

```python
new_img = cv2.resize(fix_img, (400,300))
plt.imshow(new_img)
```

```python
w_ratio = 0.5
h_ratio = 0.5

new_img2 = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
plt.imshow(new_img2)
```

```python
new_img2.shape
```

#### Flipping the image

```python
new_img3 = cv2.flip(fix_img, 0)
plt.imshow(new_img3)
```

```python
new_img3 = cv2.flip(fix_img, -1)
plt.imshow(new_img3)
```

#### Save an image

```python
cv2.imwrite('new_dog.jpg', new_img3)
```

```python
new_img4 = cv2.cvtColor(new_img3,cv2.COLOR_RGB2BGR)
cv2.imwrite('new_dog2.jpg', new_img4)
```
